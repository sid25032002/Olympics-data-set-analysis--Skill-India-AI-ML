# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
import numpy as np
df = pd.read_csv("Summer.csv")
```

### 1. In how many cities Summer Olympics is held so far?

```python
len(df['City'].unique())
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
lst=df[df['Medal']=='Gold']
data = []

for Sport in lst['Sport'].unique():
    data.append([Sport , len(lst[lst['Sport']  == Sport])])

pd.DataFrame(data,columns = ['Sport','Gold Medals']).sort_values(by='Gold Medals', ascending=False).head().plot(x = 'Sport', y = 'Gold Medals', kind = 'bar',color='black', figsize = (5,5))
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
data = []

for Sport in df['Sport'].unique():
    data.append([Sport , len(df[df['Sport']  == Sport])])

pd.DataFrame(data,columns = ['Sport','Medals']).sort_values(by='Medals', ascending=False).head().plot(x = 'Sport', y = 'Medals', kind = 'bar',color='black', figsize = (5,5))
```

### 4. Which player has won most number of medals? (Top 5)

```python
data = []

for Athlete in df['Athlete'].unique():
    data.append([Athlete , len(df[df['Athlete']  == Athlete])])

pd.DataFrame(data,columns = ['Athlete','Medals']).sort_values(by='Medals', ascending=False).head().plot(x = 'Athlete', y = 'Medals', kind = 'bar',color='black', figsize = (5,5))
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
data = []

for Event in df['Event'].unique():
    data.append([Event , len(df[df['Event']  == Event])])

pd.DataFrame(data,columns = ['Event','Players']).sort_values(by='Players', ascending=False).head().plot(x = 'Event', y = 'Players', kind = 'bar',color='black', figsize = (5,5))
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
lst=df[df['Medal']=='Gold']
mst=lst[lst['Gender']=='Women']
data = []

for Sport in mst['Sport'].unique():
    data.append([Sport , len(mst[mst['Sport']  == Sport])])

pd.DataFrame(data,columns = ['Sport','Female Gold Medalists']).sort_values(by='Female Gold Medalists', ascending=False).head().plot(x = 'Sport', y = 'Female Gold Medalists', kind = 'bar',color='black', figsize = (5,5))
```
